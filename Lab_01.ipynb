```python
people=["anna","belly","cherry","david"]
age=[21,23,34,45]
weight=[50,45,87,49]
height=[160,145,167,184]
import matplotlib.pyplot as plt
plt.scatter(age,height)
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.6,color='purple',edgecolor='k',alpha=0.75)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=16,rotation=30)
plt.yticks(fontsize=16)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='yellow',edgecolor='k',alpha=0.75,bins=5)
plt.title("histogram of patient weight",fontsize=18)
plt.xlabel("weights in kgs ",fontsize=16)
plt.xticks(fontsize=13)
plt.xticks(fontsize=15)
plt.show()
```
